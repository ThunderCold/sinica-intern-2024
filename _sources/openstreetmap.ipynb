# OpenStreetMap 作業

使用 Overpass API，在 Folium 地圖套件中載入臺北市四大超商（全家便利商店、7-Eleven、萊爾富、OK超商）之位置

## Overpass API

```python
import requests
import folium
```

```python
overpass_url = "http://overpass-api.de/api/interpreter"

overpass_query = [
"""
[out:json];
area["name:en"="Taipei"];
( 
node["name"="全家便利商店"](area);
 way["name"="全家便利商店"](area);
 rel["name"="全家便利商店"](area);
);
out center;
""",
"""
[out:json];
area["name:en"="Taipei"];
( 
node["name"="7-Eleven"](area);
 way["name"="7-Eleven"](area);
 rel["name"="7-Eleven"](area);
);
out center;
""",
"""
[out:json];
area["name:en"="Taipei"];
( 
node["name"="萊爾富"](area);
 way["name"="萊爾富"](area);
 rel["name"="萊爾富"](area);
);
out center;
""",
"""
[out:json];
area["name:en"="Taipei"];
( 
node["name"="OK超商"](area);
 way["name"="OK超商"](area);
 rel["name"="OK超商"](area);
);
out center;
"""]

response = [requests.get(overpass_url, 
                        params={'data': overpass_query[0]}),
            requests.get(overpass_url, 
                        params={'data': overpass_query[1]}),
            requests.get(overpass_url, 
                        params={'data': overpass_query[2]}),
            requests.get(overpass_url, 
                        params={'data': overpass_query[3]})]

data = [response[0].json(), response[1].json(), response[2].json(), response[3].json()]
```

將臺北市四大超商之位置標於地圖中，並且分別使用顏色區隔

- 藍色：全家便利商店
- 綠色：7-Eleven
- 紅色：萊爾富
- 白色：OK超商

```python
map = folium.Map(location=[25.06, 121.562], zoom_start=12)
color = ['blue', 'green', 'red', 'white']

for i in range(4):
    for element in data[i]['elements']:
        lat = element['lat']
        lon = element['lon']
        name = element.get('tags', {}).get('name')
    
        folium.Marker(
            location=[lat, lon],
            popup=[name,lat,lon],
            icon=folium.Icon(icon='info-sign',color=color[i]),
            tooltip='Click',
            prefix='fa'
        ).add_to(map)

map
```
